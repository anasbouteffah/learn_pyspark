### print schemas

```python
pyspark_df.printSchema()
```

```python
pyspark_df.schema
```

#### here pyspark_df = spark.read.option('header','true').csv('cheet.csv') it read all as string , to prevent this we need to add this option in read csv (inferShchem='true')

```python
pyspark_df = spark.read.option('header','true').csv('cheet.csv',inferSchema='true')
```

```python
pyspark_df.printSchema()
```

```python
type(pyspark_df)
```

```python
pyspark_df.columns
```
